```
years = df1["yıllar"].unique()
colors = ["#1f77b4", "#ff7f0e", "#2ca02c"]
plt.figure(figsize=(12, 6))
data_event = [df1[df1["yıllar"] == year][["İşgücü", "İstihdam", "İşsiz"]].values.T for year in years]
for idx, (year_data, year) in enumerate(zip(data_event, years)):
    plt.eventplot(year_data, orientation='horizontal', colors=colors, lineoffsets=idx, linelengths=0.9, linewidths=1)
plt.yticks(range(len(years)), years)
plt.xlabel("Değer")
plt.ylabel("Yıllar")
plt.title('15-24 Yaşları Arasındaki Genç Nüfusun İşgücü, İstihdam ve İşsiz Dağılımı')
plt.legend(["İşgücü", "İstihdam", "İşsiz"], title="Kategori", loc="upper right")
plt.grid(True)
plt.tight_layout()
plt.show()
```
